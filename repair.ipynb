```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder_name_lin = 'Results4_linear'
folder_name_nonlin = 'Results4_non-linear'
ligaments = ['ACLa', 'ACLp', 'PCLa', 'PCLp', 'LCL', 'MCLa', 'MCLo', 'MCLd']

dflig_lin = pd.read_csv(folder_name_lin + '/ligaments/' + ligaments[0] + '.csv', sep=',', header=None)
lig_lin = dflig_lin.to_numpy()
dflig_nonlin = pd.read_csv(folder_name_nonlin + '/ligaments/' + ligaments[0] + '.csv', sep=',', header=None)
lig_nonlin = dflig_nonlin.to_numpy()

lig_lin_force = np.empty([lig_lin.shape[0], 0])
lig_lin_length = np.empty([lig_lin.shape[0], 0])
lig_nonlin_force = np.empty([lig_nonlin.shape[0], 0])
lig_nonlin_length = np.empty([lig_nonlin.shape[0], 0])

for i in range(len(ligaments)):
    dflig_lin = pd.read_csv(folder_name_lin + '/ligaments/' + ligaments[i] + '.csv', sep=',', header=None)
    lig_lin = dflig_lin.to_numpy()
    lig_lin_force = np.append(lig_lin_force, lig_lin[:, 0].reshape(lig_lin.shape[0], 1), axis=1)
    dlength = np.empty([0])

    for j in range(lig_lin_length.shape[0]):
        dlengthj = (lig_lin[j, 5] - lig_lin[j, 4])/lig_lin[j, 4]
        dlength = np.append(dlength, dlengthj)
    lig_lin_length = np.append(lig_lin_length, dlength.reshape(lig_lin.shape[0], 1), axis=1)

    dflig_nonlin = pd.read_csv(folder_name_nonlin + '/ligaments/' + ligaments[i] + '.csv', sep=',', header=None)
    lig_nonlin = dflig_nonlin.to_numpy()
    lig_nonlin_force = np.append(lig_nonlin_force, lig_nonlin[:, 0].reshape(lig_nonlin.shape[0], 1), axis=1)
    dlength = np.empty([0])

    for j in range(lig_nonlin_length.shape[0]):
        dlengthj = (lig_nonlin[j, 5] - lig_nonlin[j, 4])/lig_nonlin[j, 4]
        dlength = np.append(dlength, dlengthj)
    lig_nonlin_length = np.append(lig_nonlin_length, dlength.reshape(lig_nonlin.shape[0], 1), axis=1)
```

```python
plt.title('ACLa')
plt.plot(lig_lin_length[:,0], lig_lin_force[:, 0], 'x-', label='lin')
plt.plot(lig_nonlin_length[:,0], lig_nonlin_force[:, 0], 'x-', label='nonlin')
plt.legend()
```
